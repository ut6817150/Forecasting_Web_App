### Creating Event type categories function

```python

def eventtype_col(df):          # This function takes in a client153 dataframe and assigns types to the EventType column
    d = {"0": "Others, Group 10",
        "12": "Others, Group 10",
        "14": "Others, Group 1",
        "15": "Others, Group 2",
        "16": "Others, Group 3",
        "17": "Others, Group 4",
        "18": "Others, Group 5",
        "19": "Others, Group 6",
        "20": "Others, Group 7",
        "21": "Others, Group 8",
        "22": "Others, Group 9",
        "23": "Others, Group 10",
        "25": "Others, Group 10",
        "26": "Others, Group 10",
        "27": "Others, Group 10",}

    df["EventType"] = df['EventId'].str[:2].map(d)
    return df
```

### Import client data and concatenate into a single dataframe

```python
dataset_filename = os.listdir(dataset_path)     # 2 client documents in path 

dtype_dict = {'EventId':'str'}

tmp_list =[]            # List of dataframes
df_count = 0            # The count of imported dataframes

client_list = ['client_153','client_219']

for fname in dataset_filename:
    for client in client_list:
        if client in fname:
            client_path = os.listdir(f"{dataset_path}")
            print(f"Reading in dataframe {df_count}: {fname}")
            client153 = eventtype_col(pd.read_csv(os.path.join(dataset_path,client_path[0]),
                                        header=0,
                                        delimiter=',',
                                        dtype = dtype_dict))
            client219 = pd.read_csv(os.path.join(dataset_path,client_path[1]),
                                                    header=0,
                                                    delimiter=',',
                                                    dtype = dtype_dict)
            
            #Append the imported dataframe to the list of dataframes
            tmp_list.append(client153)
            tmp_list.append(client219)
            #Increment the counter
            df_count += 1
            
#Concatenate the list of dataframes to a single dataframe
client_data = pd.concat(tmp_list,ignore_index = True)
```
